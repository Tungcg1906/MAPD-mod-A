```
import matplotlib.pyplot as plt
import scipy.signal as sig
import numpy as np
from math import pi
from scipy.signal import firwin, freqz
%config InlineBackend.figure_format='svg' #improve the quality of our plots
```

#Behavior of FIR filter with different numbers of taps

```
fs = 11025 #sampling frequency (Hz)
fc = 0.1*fs/2 #cutoff frequency ratio

fig = plt.figure(figsize=[12,9])
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

fig.suptitle('FIR filter response with different number of taps')

i = 0
for N, ax in zip([4,8,64, 256], [ax1, ax2, ax3, ax4]):
  
  coefs = sig.firwin(N, fc, fs=fs) #the values of our coefficients are based on this command, firwin compute the "exact values"
  [w,h] = sig.freqz(coefs) #Compute the frequency response of a digital filter
  w = fs*w/(2*pi) #convert to Hz
  h_db = 20*np.log10(abs(h))

  ax.plot(w,h_db)
  ax.vlines(fc, ymin = min(h_db), ymax = 0, colors = 'red', linestyle = '--')
  ax.set_title('N = ' + str(N))
  ax.set_xlabel('Frequency(Hz)')
  ax.set_ylabel('Amplitude(dB)')
  ax.grid('on')

  fig.subplots_adjust(hspace=0.3, wspace = 0.3)
```

#Frequency response of the FIR filter (8 taps)

```
N = 8 #taps
fs = 11025 #sampling frequency (Hz)
fc = 0.1*fs/2 #cutoff frequency ratio
print("cut freq =" ,fc)

coefs = sig.firwin(N, fc, fs=fs)
[w,h] = sig.freqz(coefs) #Compute the frequency response of a digital filter

w = fs*w/(2*pi) #convert to Hz
h_db = 20*np.log10(abs(h))


fig, ax = plt.subplots(figsize=(7,4))

ax.plot(w,h_db)
ax.vlines(fc, ymin = min(h_db), ymax = 0, colors = 'red', linestyle = '--', label = 'Cutoff frequency = %3.2f Hz' %fc)
ax.legend()
ax.set_title('Fir Filter Response')
ax.set_xlabel('Frequency(Hz)')
ax.set_ylabel('Amplitude(dB)')
ax.grid('on')
print('Coefficients = ', coefs)
```

#Definition of the FIR filter using python (8 taps)

```
N = 8 #tap
fs = 11025 #sampling frequency (Hz)
fc = 0.1*fs/2 #cutoff frequency ratio
coefs = sig.firwin(N, fc, fs=fs)

def python_fir_filter(input_data, coefs):
  filter_output = np.zeros(len(input_data))
  for n in range(len(input_data)):
      for i in range(N):
        filter_output[n] += int(coefs[i]*input_data[n-i])
  return filter_output
```
